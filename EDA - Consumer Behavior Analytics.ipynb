```python
def revrating(data):
    if data <= 2 :
        return "below 2"
    elif (data > 2) & (data <=3 ):
        return "2-3"
    elif (data > 3) & (data <=4):
        return "3-4"
    elif (data > 4) & (data <=5):
        return "4-5"
    else:
        return np.nan
```

```python
df["rating group"]= df["Review Rating"].apply(revrating)
```
